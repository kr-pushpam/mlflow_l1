```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
data_matrix= scaler.fit_transform(df)
```

```python
df
```

```python
data_matrix
```

```python
model= KMeans(n_clusters=6)
```

```python
clusters=model.fit_predict(data_matrix)
```

```python
clusters
```

```python
inertia=[]

for i in range(1,16):
  model=KMeans(n_clusters=i)
  model.fit(data_matrix)
  inertia.append(model.inertia_)

inertia
```

```python
plt.plot(range(1,16),inertia,'.-')
```

```python
model=KMeans(n_clusters=4)
clusters=model.fit_predict(data_matrix)
```

```python
clusters
```

```python
colors=['red','green','blue','yellow']
point_color=[]

for i in range(len(clusters)):
  color= colors[clusters[i]]
  point_color.append(color)
```

```python
point_color
```

```python
plt.scatter(data_matrix[:,1],data_matrix[:,2],c=point_color)
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=2)
pca.fit(data_matrix)
new_data=pca.transform(data_matrix)
```

```python
new_data
```

```python
plt.scatter(new_data[:,0],new_data[:,1],c=point_color)
```

```python
inertia=[]

for i in range(1,16):
  model=KMeans(n_clusters=i)
  model.fit(new_data)
  inertia.append(model.inertia_)

inertia
```

```python
plt.plot(range(1,16),inertia,'.-')
```

```python
model=KMeans(n_clusters=2)
clusters=model.fit_predict(new_data)
```

```python
colors=['red','green','blue','yellow']
point_color=[]

for i in range(len(clusters)):
  color= colors[clusters[i]]
  point_color.append(color)
```

```python
plt.scatter(new_data[:,0],new_data[:,1],c=point_color)
```
